2. Add Print and Append Methods

```python
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class LinkedList:
  def __init__(self, value):
    new_node = Node(value)
    self.head = new_node
    self.tail = new_node
    self.length = 1

  def append(self, value):
    new_node = Node(value)
    if self.head is None:  # when ll is empty
      self.head = new_node
      self.tail = new_node
    else:
      self.tail.next = new_node  # point current list's last node to the new node
      self.tail = new_node  # new node becomes the last node
    self.length += 1
    return True

  def print_list(self):
    ll = self.head
    while ll is not None:
      print(ll.value)
      ll = ll.next
```

```python
ll = LinkedList(1)
ll.append(2)
ll.print_list()
```

3. Add Pop Method

```python
class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class LinkedList:
  def __init__(self, value):
    new_node = Node(value)
    self.head = new_node
    self.tail = new_node
    self.length = 1

  def append(self, value):
    new_node = Node(value)
    if self.head is None:  # when ll is empty
      self.head = new_node
      self.tail = new_node
    else:
      self.tail.next = new_node  # point current list's last node to the new node
      self.tail = new_node  # new node becomes the last node
    self.length += 1
    return True

  def pop(self):
    if self.length == 0: # ll is empty
      return None
    temp = self.head
    pre = self.head  # have 2 pointers
    while temp.next:
      pre = temp
      temp = temp.next
    self.tail = pre
    self.tail.next = None
    self.length -= 1
    if self.length == 0:
      self.head = None
      self.tail = None
    return temp

  def prepend(self, value):
    new_node = Node(value)
    if self.length == 0:  # ll is empty
      self.head = new_node
      self.tail = new_node
    else:
      new_node.next = self.head
      self.head = new_node
    
    self.length += 1
    return True
  
  def pop_first(self):
    if self.length == 0:  # ll is empty
      return None
    temp = self.head
    self.head = temp.next
    temp.next = None
    self.length -= 1
    if self.length == 0:
      self.tail = None
    return temp
  
  def get(self, index):
    if index < 0 or index >= self.length:
      return None
    temp = self.head
    for _ in range(index):
      temp = temp.next
    return temp
  
  def set_value(self, index, value):
    temp = self.get(index)
    if temp:
      temp.value = value
      return True
    return False
  
  def insert(self, index, value):
    if index < 0 or index > self.length:
      return False
    if index == 0:
      return self.prepend(value)
    if index == self.length:
      return self.append(value)
    new_node = Node(value)
    temp = self.get(index - 1)
    new_node.next = temp.next
    temp.next = new_node
    self.length += 1
    return True
  
  def remove(self, index):
    if index < 0 or index >= self.length:
      return None
    if index == 0:
      self.pop_first()
    if index == self.length - 1:
      self.pop()
    prev = self.get(index - 1)
    temp = prev.next
    prev.next = temp.next
    temp.next = None
    self.length -= 1
    return temp
  
  def reverse(self):
    temp = self.head
    self.head = self.tail
    self.tail = temp
    before = None
    after = temp.next
    for _ in range(self.length):
      after = temp.next
      temp.next = before
      before = temp
      temp = after


  def print_list(self):
    ll = self.head
    while ll is not None:
      print(ll.value)
      ll = ll.next
```

```python
ll = LinkedList(0)
ll.append(1)
ll.append(2)
ll.append(3)
ll.append(4)
ll.append(5)
ll.append(6)


ll.print_list()
```

```python
ll.reverse()
ll.print_list()
```
